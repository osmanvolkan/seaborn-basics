# Scatter Plot


```python
tips.columns
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=tips.total_bill,y=tips.tip)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="smoker");
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day");
#hue kategorik data için kullanılır-evet/hayır gibi,günler gibi
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",style="sex");
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",style="sex",hue="sex");
```

```python
fig,hg=plt.subplots(1,2,figsize=(10,5))#fig yazdık başa ki her iki tabloda da parantez içine yazdığımız figsize sorunsuz çalışsın
sns.scatterplot(data=tips,x="total_bill",y="tip",style="sex",ax=hg[0]);
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",ax=hg[1]);


```

```python
#palette
plt.figure(figsize=(16,10))
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",palette="plasma");
```

```python
#heatmap
flights=sns.load_dataset("flights")
flights=flights.pivot("month","year","passengers")
ax=sns.heatmap(flights)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",alpha=0.1);
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",style="sex",hue="day",alpha=0.7);
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",style="sex",hue="day",alpha=0.7,s=100);
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",style="sex",hue="day",alpha=0.7,palette="plasma");
```

# Distribution Plot

```python
sns.displot(x="total_bill",data=tips);
sns.displot(tips["total_bill"]);
sns.displot(tips.total_bill);
```

```python
sns.displot(tips.total_bill,kde=True,bins=50);
```

```python
sns.displot(tips.tip,kde=True,bins=50);
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",s=300,hue="sex",ax=ax[0]);
sns.displot(data=tips,x="total_bill",ax=ax[1]);
                
                
                
                
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))
sns.scatterplot(data=tips,x="total_bill",y="tip",s=300,hue="sex",ax=ax[0]);    
sns.histplot(data=tips,x="tip",ax=ax[1]);    
    
    
```

```python
sns.displot(data=tips,x="total_bill",kind="hist");
sns.displot(data=tips,x="total_bill",kind="kde");
```

```python
sns.displot(data=tips,x="total_bill",kind="hist",kde=True);
```

```python
sns.displot(data=tips,x="total_bill",kind="hist",kde=True,aspect=5);
```
